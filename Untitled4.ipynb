```python
def get_odds():
    for number in range(10):
        if number % 2 != 0:
            yield number

# Part 2: Use a for loop to find and print the third value
# We can use enumerate() to count the items as we loop.
# The third item will have an index of 2 (since counting starts at 0).
for index, number in enumerate(get_odds()):
    if index == 2:  # Index 2 is the third item
        print(number)
        break       # Stop the loop once we've found it
```
